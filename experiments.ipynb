```python
from langchain_ollama import ChatOllama
from langchain_core.messages import HumanMessage

# Initialize model
llm = ChatOllama(
    model="llama3",   # make sure you pulled it: ollama pull llama3
    temperature=0
)

# Run a prompt
response = llm.invoke([
    HumanMessage(content=f"""you are an expert data engineer. 
                 Parse the script script and extract lineage information.
                 Return the lineage information in a structured format.
                 -sources
                 -targets
                 -transformations

                 Script:
                 CREATE OR REPLACE TABLE curated.customer_orders AS
                SELECT
                    c.customer_id,
                    CONCAT(c.first_name, ' ', c.last_name) AS full_name,
                    c.email,
                    o.order_id,
                    o.amount,
                    o.created_at AS order_date
                    FROM staging.customers_clean c
                    JOIN staging.orders_clean o
                    ON c.customer_id = o.customer_id;
                          )""")
                    ])

print(response.content)
```
